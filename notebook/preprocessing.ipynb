# Simple Data Preprocessing

## Import Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path="https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/main/dataset/titanic.csv"
```

## Load dataset

```python
df=pd.read_csv(path)
df.head()
```

## Cek Duplikasi Data

```python
df.duplicated().sum()
```

```python
df.drop(columns=['PassengerId'],inplace=True)
df.head()
```

```python
df.duplicated().sum()
```

## Cek Data Kosong

```python
df.isna()
```

```python
df.isna().sum()
```

```python
baris,kolom= df.shape
print(f'Jumlah data ada {baris} record, jumlah fitur ada : {kolom} kolom')
```

## Menghapus kolom yang terlalu banyak kosongnya

```python
df.drop(columns=['Cabin'],inplace=True)
df.head()
```

```python
df.sample(10)
```

```python
df[df.Age.isna()]
```

```python
df.Age.mean()
```

```python
df.Age.median()
```

```python
df.Age.mode()
```

## Mengisi data kosong dengan nilai rata-rata (numeric data field)

```python
df.Age.fillna(df.Age.mean(), inplace=True)
df.isna().sum()
```

```python
df.iloc[[5]]
```

## Melihat data yang terbanyak pada data kategori

```python
df.Embarked.value_counts()
```

```python
df[df.Embarked.isna()]
```

## Mengisi data kategori dengan data terbanyak

```python
df.Embarked.fillna('S', inplace=True)
df.isna().sum()
```
