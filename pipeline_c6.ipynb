```python
# prompt: load data dari https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/main/dataset/titanic.csv menggunakan pandas

import pandas as pd

url = "https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/main/dataset/titanic.csv"

df = pd.read_csv(url)
df.head()
```

```python
# prompt: hapus kolom passengerid, name, ticket dan cabin

df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
df.head()
```

```python
# prompt: pisahkan data fitur dan target

X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
# prompt: pisahkan data train dan test dan gunakan proporsi data kelas

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# prompt: buatkan pipeline data numerik dan kategorik

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

numeric_columns= ['Age', 'SibSp', 'Parch', 'Fare']

categorical_columns = ['Sex', 'Embarked','Pclass']


numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessing = ColumnTransformer(
    transformers=[
        ('numeric', numeric_pipeline, numeric_columns),
        ('categorical', categorical_pipeline, categorical_columns)
    ])

```

```python
# prompt: buat pipeline model dengan preprocessing sebelumnya dan model knn

from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors=3)

model_pipeline = Pipeline(steps=[
    ('preprocessing', preprocessing),
    ('model', model_knn)
])
```

```python
# prompt: fit model yang dibuat dan hitung akurasinya

model_pipeline.fit(X_train, y_train)

y_pred = model_pipeline.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.3f}')
```
