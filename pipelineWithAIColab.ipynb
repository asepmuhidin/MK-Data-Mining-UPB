```python
# prompt: with pandas load data from https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/main/dataset/titanic.csv

import pandas as pd

url = "https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/main/dataset/titanic.csv"
data = pd.read_csv(url)

# Print the first few rows of the DataFrame
data.head()
```

```python
# prompt: drop columns passegrid, name, ticket and cabin

data.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)
```

```python
# prompt: split from data to feature and target to X and y variabel

X = data.drop(columns=["Survived"])
y = data["Survived"]
```

```python
# prompt: splitting data train and test with 20% for data test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=y, random_state=42)
```

```python
X_train.head()
```

```python
# prompt: set pipeline for numeric and categorical feature

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler

num_columns = ["Age", "SibSp", "Parch", "Fare"]

numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
     ])

cate_columns = ["Sex", "Embarked","Pclass"]

cate_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("numeric", numeric_pipeline, num_columns),
        ("categorical", cate_pipeline, cate_columns),
    ]
)
```

```python
# prompt: using transformer before, build model with knn classifcation

from sklearn.neighbors import KNeighborsClassifier

model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", KNeighborsClassifier(n_neighbors=7))
])

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(f"Accuracy: {model.score(X_test, y_test)}")
```
